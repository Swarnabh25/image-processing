```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("/content/cube1.jpg",1)
img2 = cv2.imread("/content/cube2.jpg",1)
```

```python
print('Image 1 description\n')
h = img1.shape[0]
w = img1.shape[1]
c = img1.shape[2]
size1 = img1.size
dim = (h,w,c)

print('Image Dimension    : ',dim)
print('Image Height       : ',h)
print('Image Width        : ',w)
print('Number of Channels : ',c)
print('Image Size  :', size1)
```

```python
print('Image 2 description\n')
h = img2.shape[0]
w = img2.shape[1]
c = img2.shape[2]
size1 = img2.size
dim = (h,w,c)

print('Image Dimension    : ',dim)
print('Image Height       : ',h)
print('Image Width        : ',w)
print('Number of Channels : ',c)
print('Image Size  :', size1)
```

```python
img1= cv2.resize(img1, (img2.shape[1], img2.shape[0]))
```

```python
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
a = cv2.add(gray_img1, gray_img2)
plt.imshow(a)
plt.title('Add')
plt.axis('off')
plt.show()
```

```python
result1 = cv2.addWeighted(gray_img1, 0.5, gray_img2, 0.5, 0)
plt.imshow(result1)
plt.title('Add')
plt.axis('off')
plt.show()
```

```python
b = cv2.subtract(gray_img1, gray_img2)
plt.imshow(b)
plt.title('Subtract')
plt.axis('off')
plt.show()
```

```python
c = mul_cv = cv2.multiply(img1, img2)
plt.imshow(c)
plt.title('Multiplication')
plt.axis('off')
plt.show()
```

```python
h, w, channels = img2.shape
half = w//2

left_part = img2[:, :half]
right_part = img2[:, half:]

plt.imshow(left_part)
plt.axis('off')
plt.show()

plt.imshow(right_part)
plt.axis('off')
plt.show()

half2 = h//2

top = img2[:half2, :]
bottom = img2[half2:, :]

plt.imshow(top)
plt.axis('off')
plt.show()

plt.imshow(bottom)
plt.axis('off')
plt.show()
```

```python
d = cv2.bitwise_and(img2, img1)
plt.imshow(d)
plt.title('AND')
plt.axis('off')
plt.show()
```

```python
e = cv2.bitwise_or(img2, img1)
plt.imshow(e)
plt.title('OR')
plt.axis('off')
plt.show()
```

```python
f = cv2.bitwise_xor(img2, img1)
plt.imshow(f)
plt.title('XOR')
plt.axis('off')
plt.show()
```

```python
image1  = cv2.imread("/content/cube1.jpg",1)
image2  = cv2.imread("/content/cube2.jpg",1)
```

```python
# Check if images are loaded successfully
if image1 is None or image2 is None:
    print("Error: One or both images could not be loaded.")
else:
    # Ensure both images have the same dimensions
    if image1.shape == image2.shape:
        # Add the images
        added_image = cv2.add(image1, image2)

        # Plot the original and added images
        plt.figure(figsize=(10, 5))

        plt.subplot(1, 3, 1)
        plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
        plt.title('Image 1')
        plt.axis('off')

        plt.subplot(1, 3, 2)
        plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
        plt.title('Image 2')
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.imshow(cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB))
        plt.title('Added Image')
        plt.axis('off')

        plt.show()
    else:
        print("Error: Images have different dimensions. They cannot be added.")
```
