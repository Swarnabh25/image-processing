```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Scenario 1: Read grayscale digital image, Equalize Histogram, and Display Original and Equalized images

```python
# Read grayscale digital image
gray_image = cv2.imread('/content/brainpng.parspng.com-15.png', cv2.IMREAD_GRAYSCALE)

# Determine the histogram of the image
hist_original = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Equalize the histogram
equalized_image = cv2.equalizeHist(gray_image)

# Determine the histogram of the equalized image
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
```

```python
# Plot original and equalized histograms
plt.subplot(2, 1, 1)
plt.plot(hist_original, color='black')
plt.title('Original Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 1, 2)
plt.plot(hist_equalized, color='black')
plt.title('Equalized Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.show()
```

```python
# Display Original and Equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

Scenario 2: Create a binary mask, Determine histogram, and Display Original and Equalized images

```python
# Read grayscale digital image
gray_image = cv2.imread('/content/brainpng.parspng.com-15.png', cv2.IMREAD_GRAYSCALE)

# Create a binary mask of the same size as the original image
binary_mask = np.ones_like(gray_image) * 255

# Determine the histogram of the masked image
hist_masked = cv2.calcHist([gray_image], [0], binary_mask, [256], [0, 256])
```

```python
# Plot histogram of the original and masked image
plt.plot(hist_original, color='black', label='Original Histogram')
plt.plot(hist_masked, color='red', label='Masked Histogram')
plt.title('Histogram Comparison')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Display Original and Equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

Scenario 3: Read color image, Equalize Histogram, and Display Original and Equalized images

```python
# Read color image
color_image = cv2.imread('/content/brainpng.parspng.com-15.png')

# Convert color image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Determine the histogram of the image
hist_original = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Equalize the histogram
equalized_image = cv2.equalizeHist(gray_image)

# Determine the histogram of the equalized image
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
```

```python
# Plot original and equalized histograms
plt.subplot(2, 1, 1)
plt.plot(hist_original, color='black')
plt.title('Original Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 1, 2)
plt.plot(hist_equalized, color='black')
plt.title('Equalized Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.show()
```

```python
# Display Original and Equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

##Q1) A bright image will have what kind of histogram?

answer = A bright image will have a histogram skewed towards the higher end of the intensity range. This means that there will be more pixels with higher intensity values, resulting in a histogram that is shifted towards the right side.



##Q2) What is Dynamic range and Contrast of digital images?

answer = Dynamic Range in digital images refers to the range of intensity values that can be represented, typically from the darkest to the brightest. It is usually expressed in bits and determines the number of discrete intensity levels the image can have. Contrast is the difference in brightness between the lightest and darkest parts of an image. It is a measure of how sharp or distinct the edges and details are in the image.
